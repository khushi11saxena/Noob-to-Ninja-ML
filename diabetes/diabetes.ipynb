```python
import matplotlib.pyplot as plt #used for visualization of dataset
import numpy as np
import pandas as pd
import matplotlib 

df=pd.read_csv('diabetes.csv')
print("Shape of the Dataset",df.shape)
```

```python
#separating the dataset
feature= ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI','DiabetesPedigreeFunction','Age']
predicted= ['Outcome']
```

```python
X = df[feature].values
y = df[predicted].values
print("Shape of Feature and predicted column")
print(X.shape,y.shape)
```

#                                                           

### Train Test Split

```python
#Scikit-Learn is a pyhon library for machine learning in which we have set of pre defined Algorithms.
from sklearn.model_selection import train_test_split  #used for splitting the data into training and testing 
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=.2,random_state=12)

#In this i have splitted the dataset for train and test.80% data is splitted for Training and 20% data for Testing the perfrmance of our model.
```

```python
#shape of train and test datasets
Xtrain.shape,Xtest.shape,ytrain.shape,ytest.shape
```

### Data Standradization

###### It is very important to perform feature scaling here because values lies in different range.If we don't scale the features then it maybe possible that do different features might dominate others feature when the model finds the nearest neighbor to a data point in the data space.                                                                                                            

```python
from sklearn.preprocessing import StandardScaler  #used for Standtazing the data
std = StandardScaler()
```

```python
std_x = std.fit_transform(X)
Xtrain = std.fit_transform(Xtrain) 
Xtest = std.transform(Xtest)
print("Stadraized Data ",std_x)
```

 #

### Training Data Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression #algorithm for training the data


log=LogisticRegression(random_state=0)
log.fit(Xtrain,ytrain)
```

#####     

##### Decision Tree Classification

######
 Using Decision Tree Clasification for easy visualization of Data

```python
#Using Temporary values for Making Decision Tree
x2train=Xtrain
x2test=Xtest
y2train=ytrain
y2test=ytest
```

```python
from sklearn.tree import DecisionTreeClassifier #algorithm used for making decision tree classifier

dtc= DecisionTreeClassifier()

dtc.fit(x2train,y2train)

test_score2= dtc.score(x2test,y2test)

print('test score',test_score2)

train_score2 = dtc.score(x2train,y2train)

print('train score',train_score2)
```

```python
from sklearn import tree #used for plotting the tree

dc=DecisionTreeClassifier(criterion="gini",max_depth=4)
dc=dtc.fit(x2train,y2train)
tree.plot_tree(dc)
```

##### End of Decision Tree

#    

### Accuracy Score 


```python
from sklearn.metrics import accuracy_score #used for finding the accuracy score
```

```python
#Accuracy Score of Training Data
x_pred1 = log.predict(Xtrain)
train_accuracy=accuracy_score(x_pred1,ytrain)
print("Accuracy Score of Train Data",train_accuracy)
```

```python
#Accuracy Score of Testing Data
x_pred2 = log.predict(Xtest)
test_accuracy=accuracy_score(x_pred2,ytest)
print("Accuracy Score of Test Data",test_accuracy)
```

### Prediction System


##### Choose the data from CSV file ( diabetes.csv ) for finding the prediction

```python
#Put the input data using the CSV file

input_data=(10,168,74,0,0,38,0.537,34)


#input data as array 
input_array=np.asarray(input_data)

#reshape array for only one instance
input_reshape=input_array.reshape(1,-1)

#Standradize data 
std_input=std.transform(input_reshape)
print("Standradize Input Data ",std_input)
```

```python
pred=log.predict(std_input)
print("Prediction For Diabetic Person:",pred)
```
